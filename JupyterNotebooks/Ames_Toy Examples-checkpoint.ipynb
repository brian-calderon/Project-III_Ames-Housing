### RF feature importance

```python
from sklearn import ensemble
from sklearn.model_selection import GridSearchCV, KFold, StratifiedKFold, train_test_split
```

```python
# Create the object for the RF Regressor
RF = ensemble.RandomForestRegressor()

# Set the parameters of your RF regressor. 
# max_features = Each tree gets the full set of features, but at each node, 
#                only a random subset of features is considered. 
#                https://sebastianraschka.com/faq/docs/random-forest-feature-subsets.html
#                Sci-kit reccomends m=p for regression problems & m=sqrt(p) for classification
#                You should always tune m regardless of what scikit reccomends (see below link)
#                https://stats.stackexchange.com/questions/324370/references-on-number-of-features-to-use-in-random-forest-regression
# n_estimators: # trees in forest
RF.set_params(random_state = 0, n_estimators=30, max_features=2)

# fit the RF model to the traiing data
RF.fit(X_train, y_train)
# get the model scor
RF.score(X_train, y_train)
```

```python
# create a zipped tuple of training column names and feature importance score.
feature_importance = list(zip(X_train.columns, RF.feature_importances_))
# define numpy data type for each element in tuple. 
# S10: string of size 10
# float: self explanatory
dtype = [('feature', 'S10'), ('importance', 'float')]
# Create a numpy structured array: https://numpy.org/doc/stable/user/basics.rec.html
feature_importance = np.array(feature_importance, dtype=dtype)
# [::-1] reverse order of array
feature_sort = np.sort(feature_importance, order='importance')[::-1]
# unpack the zipped array into column names and scores
name, score = zip(*feature_sort)
pd.DataFrame({'name':name,'score':score})[:15].plot.bar(x='name', y='score')
```

### RF vs MLR

```python
# RF + GridsearchCV
np.random.seed(0)
grid = {'random_state': [0], 'n_estimators': range(30,300,30), 'max_features': range(1,4,1)} #, 'max_depth':range(3,20,2)}
# intialize the grid search.
# CV = Kfold cross validation. This says how many subgroups to divide up your test data
RF_gsCV = GridSearchCV(ensemble.RandomForestRegressor(), grid, cv=2, return_train_score=True, n_jobs=4)
RF_gsCV.fit(X_train, y_train)
```

```python
print(RF_gsCV.best_params_)
RF_gsCV.best_score_
# RF_gsCV.cv_results_
```

```python
# Create the object for the RF Regressor
RF2 = ensemble.RandomForestRegressor()

# Set the parameters of your RF regressor. 
# max_features = Each tree gets the full set of features, but at each node, 
#                only a random subset of features is considered. 
#                https://sebastianraschka.com/faq/docs/random-forest-feature-subsets.html
#                Sci-kit reccomends m=p for regression problems & m=sqrt(p) for classification
#                You should always tune m regardless of what scikit reccomends (see below link)
#                https://stats.stackexchange.com/questions/324370/references-on-number-of-features-to-use-in-random-forest-regression
# n_estimators: # trees in forest
RF2.set_params(random_state = 0, n_estimators=150, max_features=1)

# fit the RF model to the traiing data
RF2.fit(X_train, y_train)
# get the model scor
RF2.score(X_train, y_train)
```

```python
from sklearn.linear_model import ElasticNet
# elasticnet = ElasticNet(alpha=0.1, l1_ratio=0.5, normalize=False)
elasticgrid = {'alpha': range(0,100,1), 'l1_ratio': np.linspace(0,1,11), 'normalize': [True]}
elasticCV = GridSearchCV(ElasticNet(), elasticgrid, cv = 3, return_train_score = True, n_jobs = 4)
elasticCV.fit(X_train,y_train)
```

```python
print(elasticCV.best_params_)
elasticCV.best_score_
# elasticCV.coef_
```

```python
elastic = ElasticNet(alpha = 16, l1_ratio = 1.0, normalize = True)
elastic.fit(X_train, y_train)
print(elastic.score(X_train, y_train))
```

```python
# create a zipped tuple of training column names and feature importance score.
elastic_FI = list(zip(X_train.columns, abs(elastic.coef_)))
# define numpy data type for each element in tuple. 
# S10: string of size 10
# float: self explanatory
dtype = [('feature', 'S10'), ('importance', 'float')]
# Create a numpy structured array: https://numpy.org/doc/stable/user/basics.rec.html
elastic_FI = np.array(elastic_FI, dtype=dtype)
# [::-1] reverse order of array
elasticsort_FI = np.sort(elastic_FI, order='importance')[::-1]
# unpack the zipped array into column names and scores
name, score = zip(*elasticsort_FI)
pd.DataFrame({'name':name,'score':score})[:15].plot.bar(x='name', y='score')

# The values here are the coeff. for the linear elastic model, these don't neccesarily
# predict significance, You can hava a large coeff. for bad normalization. Also, these
# coefficients are how the target varies with the rest held constant, therefore not really 
# a significance amongst all inputs. P_values for these models would be better, but they 
# are not available yet. Do not infer feature importance from the below histogram.
```
